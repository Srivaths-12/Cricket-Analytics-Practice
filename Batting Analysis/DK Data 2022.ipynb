```python
import math
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)

import matplotlib.pyplot as plt

deliveries = pd.read_csv('deliveries_updated_mens_ipl.csv')
matches = pd.read_csv('matches_updated_mens_ipl.csv')

df = deliveries.copy()
mdf = matches.copy()
```

```python
req_df = pd.merge(df,mdf, on = 'matchId', how = 'left')
```

```python
def phase(over):
    if over < 6:
        return 'Powerplay Overs'
    elif over < 16:
        return 'Middle Overs'
    else:
        return 'Death Overs'
```

```python
req_df['Phase'] = req_df['over'].apply(lambda x : phase(x))
```
